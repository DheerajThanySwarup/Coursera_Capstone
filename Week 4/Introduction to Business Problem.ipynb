# The Battle of Neighbourhood - Week 1

## Business Proposal

### Introdcution:

A lot of people are migrating to canada as the country offers job opportunities in various sector and also has been a major attraction for tourists. It is one of the most sought-after countries for people planning to move abroad. The purpose of this project is to suggest a better neighbourhood for people migrating to Toronto. A best neighbourhood can be defined as a place which have an ease of accessing to School, Hospital, Restaurant, Gym, Movie Theatre, etc.


### Data Description:

This project would use Four-square API as the data source and the API provides the ability to perform location search, and location sharing. We will use the API to collect the information about neighbourhood surroundings.

### Methodology:

We will be using unsupervised machine learning algorithm "k-means clustering algorithm" to explore the neighbourhood, segment them and group them into clusters.

### Libraries:

Pandas: For data wrangling.

Folium: To visulaize map with clusters of neighbourhoods.

Scikit Learn: For clustering algorithm.

Geocoder: To retrieve location data.

Requests: To handle HTTP requests.

Matplotlib: For plotting.

### Stakeholders:

Canada permanent residency VISA holders planning to migrate to Toronto.
